```python
#quick diagram of activation functions compared to each other
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

```python
AFs=[nn.ReLU(), nn.LeakyReLU() ,nn.PReLU(), nn.Tanh(), nn.Mish(), nn.RReLU(), nn.SELU(), nn.ELU(), nn.Sigmoid(), nn.Softplus()]
```

```python
space=torch.linspace(-2, 2, 100)
with torch.no_grad():
    plt.figure(figsize=(10, 7))
    for i in range(0,5):
        plt.plot(space,AFs[i](space), alpha=.75)
    plt.title('Activation function comparisons [1]')
    plt.legend(AFs[:5])
    plt.savefig('./AF1.png')
    plt.show()
    plt.figure(figsize=(10, 7))
    for i in range(5,10):
        plt.plot(space,AFs[i](space), alpha=.75)
        plt.title('Activation function comparisons [2]')
    plt.legend(AFs[5:])
    plt.savefig('./AF2.png')
    plt.show()
```
